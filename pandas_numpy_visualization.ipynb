## `Sort_values()` Method
- The sort_values method sorts a Series values in order.
- By default, pandas applies an ascending sort (smallest to largest).
- Customize the sort order with the ascending paramete

| Parameter      | What it Takes | What it Does |
|----------------|--------------|--------------|
| **by**         | String (column name) or list of strings | Specifies the column(s) to sort by. |
| **axis**       | `0` or `1` (default `0`) | `0` → sort rows, `1` → sort columns. |
| **ascending**  | Boolean or list of booleans (default `True`) | `True` → ascending order, `False` → descending order. For multiple columns, provide a list of booleans. |
| **inplace**    | Boolean (default `False`) | If `True`, sorts the DataFrame in place without returning a new object. |
| **na_position**| `'first'` or `'last'` (default `'last'`) | Specifies whether NaN values appear at the beginning or the end. |
| **ignore_index** | Boolean (default `False`) | If `True`, resets the index in the sorted result to 0, 1, 2, ... |
| **key**        | Function (default `None`) | A function applied to each column before sorting, useful for custom sorting logic (Pandas 1.1+). |

## `dropna()` method -  Drop Rows with Missing Values
- Pandas uses a `NaN` designation for cells that have a missing value.
- The `dropna` method deletes rows with missing values. Its default behavior is to remove a row if it has *any* missing values.
- Pass the `how` parameter an argument of "all" to delete rows where all the values are `NaN`.
- The `subset` parameters customizes/limits the columns that pandas will use to drop rows with missing values.

| Parameter      | What it Takes | What it Does |
|----------------|--------------|--------------|
| **axis**       | `0` or `1` (default `0`) | `0` → drop rows with missing values, `1` → drop columns with missing values. |
| **how**        | `'any'` or `'all'` (default `'any'`) | `'any'` → drop if **any** value is NaN, `'all'` → drop only if **all** values are NaN. |
| **thresh**     | Integer (default `None`) | Minimum number of **non-NaN** values required to keep the row/column. Overrides `how` if set. |
| **subset**     | Label or sequence of labels (default `None`) | Specifies which columns (axis=0) or rows (axis=1) to check for missing values. |
| **inplace**    | Boolean (default `False`) | If `True`, modifies the DataFrame in place without returning a new object. |
| **ignore_index** | Boolean (default `False`) | If `True`, resets the index in the result to 0, 1, 2, ... after dropping rows/columns. |

## `fillna()` method - Fill in Missing Values with the fillna Method

| Parameter      | What it Takes | What it Does |

| **inplace**    | Boolean (default `False`) | If `True`, modifies the DataFrame/Series in place without returning a new object. |
| **limit**      | Integer (default `None`) | Maximum number of consecutive NaNs to fill. |
| **downcast**   | Dict or string (default `None`) | Attempt to downcast filled data to a smaller dtype (e.g., `int64` → `int32`) when possible. |

## `sort_values()` method

- The `sort_values` method sorts a **DataFrame** by the values in one or more columns. The default sort is an ascending one (alphabetical for strings).
- The first parameter (`by`) expects the column(s) to sort by.
- If sorting by a single column, pass a string with its name.
- The `ascending` parameter customizes the sort order.
- The `na_position` parameter customizes where pandas places `NaN` values.
- To sort by multiple columns, pass the `by` parameter a list of column names. Pandas will sort in the specified column order (first to last).
- Pass the `ascending` parameter a Boolean to sort all columns in a consistent order (all ascending or all descending).
- Pass `ascending` a list to customize the sort order *per* column. The `ascending` list length must match the `by` list.

| Parameter      | What it Takes | What it Does |
|----------------|--------------|--------------|
| **by**         | Label or list of labels | Column(s) or row index name(s) to sort by. Required for DataFrame; ignored for Series. |
| **axis**       | `0` or `1` (default `0`) | `0` → sort rows by column values, `1` → sort columns by row values. |
| **ascending**  | Boolean or list of booleans (default `True`) | `True` → ascending order, `False` → descending order. For multiple sort keys, provide a list. |
| **inplace**    | Boolean (default `False`) | If `True`, performs sorting in place without returning a new object. |
| **kind**       | `'quicksort'`, `'mergesort'`, `'heapsort'`, `'stable'` (default `'quicksort'`) | Sorting algorithm to use. `'mergesort'` and `'stable'` preserve the order of equal elements. |
| **na_position**| `'first'` or `'last'` (default `'last'`) | Specifies whether NaN values appear at the beginning or the end. |
| **ignore_index** | Boolean (default `False`) | If `True`, resets the index in the sorted result to 0, 1, 2, ... |
| **key**        | Callable (default `None`) | Function applied to values before sorting. Useful for custom sorting logic. |

```python
df.head(2)
```

```python
df.sort_values(by="First Name", ascending=False, na_position="first")
```

## rank() method
- The rank method assigns a numeric ranking to each Series value.
- Pandas will assign the same rank to equal values and create a "gap" in the dataset for the ranks.

| Parameter      | What it Takes | What it Does |
|----------------|--------------|--------------|
| **axis**       | `0` or `1` (default `0`) | `0` → rank over rows (column-wise), `1` → rank over columns (row-wise). |

## The pivot Method
- The `pivot` method reshapes data from a tall format to a wide format.
- Ask yourself which direction the data will expand in if you add more entries.
- A tall/long format expands down. A wide format expands out.
- The `index` parameter sets the horizontal index of the pivoted **DataFrame**.
- The `columns` parameter sets the column whose values will be the columns in the pivoted **DataFrame**.
- The `values` parameter set the values of the pivoted **DataFrame**. Pandas will populate the correct values based on the index and column intersections.

| Parameter      | What it Takes | What it Does |
|----------------|--------------|--------------|
| **index**      | str or object | Column to use to make new frame’s index. |
| **columns**    | str or object | Column to use to make new frame’s columns. |
| **values**     | str, object, or list of them, optional | Column(s) to use for populating new frame’s values. If not specified, all remaining columns are used. |

## The pivot_table Method
- The `pivot_table` method operates similarly to the Pivot Table feature in Excel.
- A pivot table is a table whose values are aggregations of groups of values from another table.
- The `values` parameter accepts the numeric column whose values will be aggregated.
- The `aggfunc` parameter declares the aggregation function (the default is mean/average).
- The `index` parameter sets the index labels of the pivot table. MultiIndexes are permitted.
- The `columns` parameter sets the column labels of the pivot table. MultiIndexes are permitted.

| Parameter      | What it Takes | What it Does |
|----------------|--------------|--------------|
| **data**       | DataFrame | The DataFrame to pivot. |
| **values**     | str, object, or list of them, optional | Column(s) whose values will populate the table cells. If not specified, all remaining columns are used. |
| **index**      | str, object, or list of them, optional | Column(s) to set as the table’s index (rows). |
| **columns**    | str, object, or list of them, optional | Column(s) to set as the table’s columns. |
| **aggfunc**    | Function, list, dict, or default `mean` | Aggregation function(s) to apply to grouped data (e.g., `mean`, `sum`, `min`, `max`). |
| **fill_value** | Scalar (default `None`) | Value to replace missing values in the resulting table. |
| **margins**    | Boolean (default `False`) | If `True`, adds special rows/columns with all values (e.g., totals). |
| **margins_name** | str (default `'All'`) | Name for the totals row/column when `margins=True`. |
| **dropna**     | Boolean (default `True`) | If `True`, does not include columns whose entries are all NaN. |
| **observed**   | Boolean (default `False`) | For categorical groupers, if `True`, only observed category combinations appear. |
| **sort**       | Boolean (default `True`) | If `True`, sorts the table’s index and columns. |
