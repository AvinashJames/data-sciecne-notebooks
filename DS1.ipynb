### CHI Square test with Python

```python
import scipy.stats as stat
observed_val = [12,34,56,78,65,34]
expected_val = [11,32,55,70,63,48]

chi2_stat, p_value = stat.chisquare(f_obs=observed_val,f_exp=expected_val)
print(f"The chi2stat is {chi2_stat} and p_value is {p_value}")

if p_value <=0.05:
    print("we will reject the Null Hypothesis")
else:
    print("We will fail to reject the Null Hypothesis")
```

```python
# finding the critical value
significance_value = 0.05
dof = len(observed_val)-1
print(dof)
critical_value = stat.chi2.ppf(1-significance_value,dof)
```

## Handling Imbalanced Dataset
##### 1. Upsampling
##### 2. Downsampling

```python
# Creating the dataset
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)  # setting random seed for reproducibility

# Create a dataframe with two features
n_sample = 1000
class_0_ratio = 0.9
n_class_0 = int(n_sample * class_0_ratio)
n_class_1 = n_sample - n_class_0
```

```python
n_class_0
```

```python
n_class_1
```

```python
class_0 = pd.DataFrame({
    "featre_1":np.random.normal(loc=0.0, scale=1.0, size=n_class_0),
    "feature_2":np.random.normal(loc=0.0, scale = 1.0, size = n_class_0),
    "target":[0] * n_class_0
})

class_1 = pd.DataFrame({
    "featre_1":np.random.normal(loc=2.0, scale=1.0, size=n_class_1),
    "feature_2":np.random.normal(loc=2.0, scale = 1.0, size = n_class_1),
    "target":[1] * n_class_1
})
```

```python
class_0
```

```python
class_1
```

```python
final_df = pd.concat([class_0, class_1],ignore_index=True)
final_df
```

```python
final_df["target"].value_counts() #This clearly gives the imbalanced dataset
```

#### Upsampling

```python
df_majority = final_df[final_df["target"]==0]
```

```python
df_majority.info()
```

```python
df_minoroty = final_df[final_df["target"]==1]
```

```python
df_minoroty.info()
```

```python
minority_upsampling.info()
```

```python
# Concating the minority_upsample_df with df_majoirty
final_upsampled_df = pd.concat([df_majority,minority_upsampling],ignore_index=True)
```

```python
import numpy as np
lst_marks = [45,22,34,65,87,98,57,48,29,999,1100]
```

```python
val1.min()
```

#### Lower Fence ----  Higher Fence

```python
## Anything in between the lower fence and higher fence is good.
## Anything which is out of this fence is outlier.
Q1 = np.percentile(lst_marks,[25])
```

```python
Q1
```

```python
Q3 = np.percentile(lst_marks,[75])
Q3
```

```python
# np quantile is used to give the data about the min,Q2,Q3,Q4 and max
min,Q1,Q2,Q3,max = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
min
```

```python
Q1,Q3
```

```python
# IQR is important for getting the lower fence and higher fence
IQR = Q3-Q1
IQR
```

```python
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
```

```python
lower_fence,higher_fence
```

```python
lst_marks
new_marks = []
outlier_marks = []
for i in lst_marks:
    if i >= lower_fence and i <= higher_fence:
        new_marks.append(i)
    else:
        outlier_marks.append(i)
        
        
```

```python
new_marks,outlier_marks
```

```python
import seaborn as sns
sns.boxenplot(lst_marks)
# we can get the outliers clearly in the plot
```

```python
sns.boxplot(new_marks) # no outliers is present here
```
